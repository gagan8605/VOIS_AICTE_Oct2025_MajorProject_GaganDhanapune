```python
def split_explode(df, col, sep=','):
    """Return exploded DataFrame for a multi-valued column."""
    s = df[col].fillna('').astype(str).str.split(sep)
    s = s.apply(lambda x: [i.strip() for i in x if i and i.strip()!=''])
    exploded = df.loc[s.index].copy()
    exploded = exploded.assign(**{col: s})
    exploded = exploded.explode(col)
    exploded[col] = exploded[col].replace('', np.nan)
    return exploded
```

```python
df_genres = split_explode(df, 'type')
df_countries = split_explode(df, 'country')
```

```python
import re
def parse_duration(d):
    if pd.isna(d): return np.nan
    d = str(d)
    m = re.search(r'(\d+)\s*min', d, re.I)
    if m: return int(m.group(1))
    else:
        # for TV shows, try seasons
        s = re.search(r'(\d+)\s*Season', d, re.I)
        if s: return int(s.group(1))
    return np.nan

df['duration_val'] = df['duration'].apply(parse_duration)
```

```python
import plotly.express as px

# Get the value counts for countries
country_counts = df_countries['country'].value_counts().reset_index()
country_counts.columns = ['country', 'count']

# Generate the treemap
fig = px.treemap(country_counts.head(20), path=['country'], values='count',
                 title='Distribution of Content by Top 20 Countries')
fig.show()
```

```python
from matplotlib.ticker import PercentFormatter

country_counts = df_countries['country'].value_counts()

# Calculate cumulative percentage
cumulative_percentage = country_counts.cumsum() / country_counts.sum() * 100

fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for individual counts
ax1.bar(country_counts.index[:20], country_counts.values[:20])
ax1.set_xlabel('Country')
ax1.set_ylabel('Count', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(country_counts.index[:20], rotation=45, ha='right')

# Line chart for cumulative percentage
ax2 = ax1.twinx()
ax2.plot(cumulative_percentage.index[:20], cumulative_percentage.values[:20], color='red', marker='o')
ax2.set_ylabel('Cumulative Percentage', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.yaxis.set_major_formatter(PercentFormatter())

plt.title('Pareto Chart of Content by Top 20 Countries')
plt.show()
```
